## using multiple linear regression analysis

spliting data

```python
from sklearn.model_selection import train_test_split
```

```python
Y=df['House price of unit area']
```

```python
X=df[['House Age','Distance from nearest Metro station (km)','Number of convenience stores']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

training model

```python
from sklearn.linear_model import LinearRegression
```

```python
Lr=LinearRegression()
```

```python
Lr.fit(x_train,y_train)
```

```python
y_predictions=Lr.predict(x_test)
```

```python
y_predictions
```

accuracy and errors

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
score=r2_score(y_test,y_predictions)
mse=mean_squared_error(y_test,y_predictions)
rmse=np.sqrt(mse);
```

```python
print("r2 score is : ",score)
print("mse is:",mse)
print("rmse is:",rmse)
```

## using random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
random_forest = RandomForestRegressor(n_estimators=100)
random_forest.fit(x_train, y_train)
y_predictions1 = random_forest.predict(x_test)
```

```python
score1=r2_score(y_test,y_predictions1)
mse1=mean_squared_error(y_test,y_predictions1)
rmse1=np.sqrt(mse1);
```

```python
print("r2 score is : ",score1)
print("mse is:",mse1)
print("rmse is:",rmse1)
```

## using support vector machines 

```python
from sklearn.svm import SVR
```

```python
svr = SVR(C=100000)
svr.fit(x_train, y_train)
y_predictions2 = svr.predict(x_test)
```

```python
score2=r2_score(y_test,y_predictions2)
mse2=mean_squared_error(y_test,y_predictions2)
rmse2=np.sqrt(mse2);
```

```python
print("r2 score is : ",score2)
print("mse is:",mse2)
print("rmse is:",rmse2)
```

```python
x_labels=['linear regression','randomforest','SVM']
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x=x_labels, y=[rmse,rmse1,rmse2],width=0.5)
plt.title("Models' RMSE Scores (Cross-Validated)", size=10)
plt.xticks(rotation=30, size=15)
plt.yticks(ticks=np.arange(0, 8, step=0.3),rotation=30, size=15)
plt.show()
```
